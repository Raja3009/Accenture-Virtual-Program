```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("Documents/Reactions.csv",index_col = 0)
```

```python
df1
```

```python
df2 = pd.read_csv('Documents/ReactionTypes.csv',index_col = 0)
```

```python
df2 
```

```python
df3 = pd.read_csv('Documents/Content.csv',index_col=0)
```

```python
df3
```

```python
df1.shape #Reactions
```

```python
df2.shape #Reactiontypes
```

```python
df3.shape #Content
```

```python
df4 = df1.merge(df3,how = 'left',on =['Content ID'])
```

```python
df4
```

```python
df4.shape
```

```python
df4.drop('User ID_x',axis = 1,inplace = True)
```

```python
df4
```

```python
df4.rename(columns = {'Type_x':'Reaction_Type','Type_y':'Content_type','User ID_y':'User ID'},inplace = True)
```

```python
df4
```

```python
df5 = df4.merge(df2,how = 'left',left_on='Reaction_Type',right_on='Type')
```

```python
df5.drop('Type',inplace = True,axis = 1)
```

```python
df5
```

```python
df5.drop('URL',inplace = True,axis =1 )
```

```python
df5.shape
```

```python
finalised_dataset = df5.dropna(subset=['Score']) #dropping rows where values are missing
```

```python
finalised_dataset.shape
```

```python
finalised_dataset
```

```python
finalised_dataset['Category'].unique()
```

```python
finalised_dataset['Category'] = finalised_dataset['Category'].str.upper()
```

```python
finalised_dataset['Category'].unique()
```

```python
finalised_dataset['Category'] = finalised_dataset['Category'].str.replace('"', '')
finalised_dataset['Category'].unique()
```

```python
finalised_dataset['Category'].nunique()
```

```python
finalised_dataset['Category'].value_counts().head()
```

```python
finalised_dataset
```

```python
finalised_dataset.to_csv('Finalised_dataset.csv')
```

```python
finalised_dataset.groupby('Category').sum().sort_values(by = 'Score',ascending=False).head(5)
```

```python
case_study = finalised_dataset.groupby('Category').sum().sort_values(by = 'Score',ascending=False).head(5)
```

```python
type(case_study)
```

```python
case_study.columns
```

```python
case_study.reset_index(inplace= True)
```

```python
case_study
```

```python
case_study.columns
```

```python
category = case_study['Category']
category
```

```python
x = list(category)
```

```python
score = case_study['Score']
```

```python
score
```

```python
y = list(score)
```

```python
xpos = np.arange(len(x))
xpos
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('Sum of Score in each category')
plt.ylabel("Category")
plt.title('Case study of top 5 categories')
plt.yticks(xpos,x)
plt.barh(xpos,y,height=0.5,color = 'green')
plt.savefig('Barchart.png',bbox_inches="tight", pad_inches=1, transparent=True)
plt.show()
```

```python
plt.pie(y,labels=x,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig('Pie_chart.png',bbox_inches="tight", pad_inches=1, transparent=True)
plt.show()
```

```python
import seaborn as sns
plt.title('Popularity Score by Category')
plt.xlabel('Category')
plt.ylabel('Aggregates of Score')
plt.plot(x,y,marker = '*',markerfacecolor = 'Green')
plt.savefig('Linechart.png')
```
